```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from numpy.linalg import inv

data =np.array([[1.5, 2], [2, 2.5], [2.5, 3]])
X, y = data[:,0], data[:,1]

X_mean = np.mean(X)
y_mean = np.mean(y)
print('X mean :',X_mean," y_mean :",y_mean)
covariance = np.sum((X-X_mean)*(y-y_mean))
variance = np.sum((X-X_mean)*(X-X_mean))
print('Covariance :',covariance, " Varinace: ",variance)
theta1 = covariance/variance
print("theta1: ",theta1)
theta0 = y_mean - theta1 * X_mean
print("theta0: ",theta0)
SSE = np.sum((y-(theta1*X + theta0))*(y-(theta1*X + theta0)))
print("SSE: ",SSE)

plt.scatter(X,y,color='blue')
plt.scatter(X, y)
plt.title('Data Scatter Diagram')
plt.xlabel('Monthly Salary')
plt.ylabel('Loan Received')
plt.show()

```

```python
# We suppose theta0= 0
# scenario 1
theta1_1 = 0.5
print("theta1_1: ", theta1_1)
# predict using coefficients
yhat_1 = X.dot(theta1_1)
# plot predictions
plt.scatter(X, y)
plt.title('Data Scatter Diagram')
plt.xlabel('Monthly Salary')
plt.ylabel('Loan Received')
plt.plot(X, yhat_1,label='theta = 0.5', color='red')

#senario 2
theta1_2 = 1
print("theta1_2: ", theta1_2)
# predict using coefficients
yhat_2 = X.dot(theta1_2)
# plot predictions
plt.plot(X, yhat_2,label='theta = 1.0', color='brown')

#senario 3
theta1_3 = 1.5
print("theta1_3: ", theta1_3)
# predict using coefficients
yhat_3 = X.dot(theta1_3)
# plot predictions
plt.plot(X, yhat_3,label='theta = 1.5', color='purple')

# Optimal senario
# We suppose theta0= 0 and theta1 if calculated
X = X.reshape((len(X), 1))
theta1 = inv(X.T.dot(X)).dot(X.T).dot(y)
print("theta1: ", theta1)
# predict using coefficients
yhat = X.dot(theta1)
# plot data and predictions
plt.plot(X, yhat, label = 'Optimal', color='pink')

plt.legend()
plt.show()
```
